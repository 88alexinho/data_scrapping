```python
import requests
from bs4 import BeautifulSoup
import lxml
import pandas as pd
```

```python
url = 'https://www.rbc.ru/life/news/65570bc99a79472c4f509a95'
```

```python
r = requests.get(url)
r.text
```

```python
soup = BeautifulSoup(r.text,'lxml')
```

```python
link = 'https://www.rbc.ru' + soup.find('h2', class_='subtitle secondary').find('a',class_='link primary').get('href')
link
```

```python
russian_name = soup.find('h2', class_='subtitle secondary').text.split('.')[1].strip('()').split('» (')[0]
year = soup.find('h2', class_='subtitle secondary').text.split(' ')[-1].strip('()')
russian_name, year
```

```python
name = soup.find('h2', class_='subtitle secondary').text.split(' ')
name
```

```python
films = soup.findAll('h2',class_='subtitle secondary')
films
```

```python
len(films)
```

```python
films = soup.findAll('h2',class_='subtitle secondary')

data = []

for film in films:
    russian_name = film.text.split('.')[1].strip('()').split('» (')[0].split('«')[1]
    year = film.text.split(' ')[-1].strip('()')

    data.append([year, russian_name])
```

```python
data
```

```python
USSR_films = pd.DataFrame(data,columns=['production_year','film_name'])
```
